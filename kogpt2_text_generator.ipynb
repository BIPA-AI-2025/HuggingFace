## skt/kogpt2-base-v2 같은 KoGPT-2로 단어 생성 

```python
from transformers import pipeline
```

```python
# 1) 파이프라인 생성
gen_ko = pipeline(
    "text-generation",
    model="skt/kogpt2-base-v2",
    tokenizer="skt/kogpt2-base-v2",
    framework="pt",            # PyTorch 백엔드 사용
    device=-1,                 # CPU 사용(여기에 GPU 인덱스를 넣으면 GPU 사용)
    max_length=12,             # 프롬프트 포함 최대 토큰 길이
    do_sample=True,            # 샘플링 방식으로 생성
    top_k=50,                  # 탑 k 샘플링
    top_p=0.95,                # 누적 확률 p 컷오프
    repetition_penalty=1.2,    # 반복 패널티
    max_new_tokens=3,      # 프롬프트 이후 토큰만 3개 생성
)
```

```python
# 2) 단어(신조어) 생성 함수
def generate_korean_word(prompt, num_return_sequences=5):
    """
    prompt: 생성 시작 프롬프트 (예: "신조어: ")
    num_return_sequences: 몇 개의 후보를 뽑을지
    """
    outputs = gen_ko(
        prompt,
        num_return_sequences=num_return_sequences
    )
    # 결과 텍스트에서 프롬프트 부분을 제거하고 단어만 리턴
    words = []
    for out in outputs:
        text = out["generated_text"]
        word = text[len(prompt):].split()[0].strip()
        words.append(word)
    return words
```

```python
# 3) 바로 실행
new_words = generate_korean_word(prompt="한 단어만 만들어줘: ", num_return_sequences=5)
for idx, w in enumerate(new_words, 1):
    print(f"{idx}. {w}")
```
