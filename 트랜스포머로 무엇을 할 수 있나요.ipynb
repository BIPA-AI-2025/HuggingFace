## NER(Named Entity Recognition)

* NER는 문장에서 사람 이름, 장소, 조직 이름 같은 "특별한 이름(엔터티)"들을 자동으로 찾아내는 작업

* dbmdz/bert-large-cased-finetuned-conll03-english 

  Hugging Face : https://huggingface.co/dbmdz/bert-large-cased-finetuned-conll03-english

- Entity

  - PER: Person (인물 이름)

  - ORG: Organization (회사, 기관 등)

  - LOC: Location (지리적 장소)

  - start ~ end (단어의 위치)

```python
from transformers import pipeline

ner = pipeline("ner", aggregation_strategy="simple")
ner("My name is Sylvain and I work at Hugging Face in Brooklyn.")
```

```python
# 한국어  NER 전용 모델이 아니기 떄문에 [] 출력
ner("안녕하세요. 오늘코드 유튜브 채널입니다. 여기는 대한민국입니다.")
```

```python
ner("Hello, I'm Korean")
```

### 한국어 NER 모델 : kykim/bert-kor-base

```python
# kykim/bert-kor-base는 NER로 파인튜닝되지 않은 base 모델 → 분류 헤드가 랜덤 초기화되어 결과가 무의미함.
# 실제 한국어 NER 파인튜닝 모델로 교체 (PS:인물, LC:지명, DT:시간, OG:기관, CV:문명/직위 등)
ner = pipeline(
    "ner",
    model="Leo97/KoELECTRA-small-v3-modu-ner",
    aggregation_strategy="simple"
)

text = "안녕하세요. 오늘코드 유튜브 채널입니다. 여기는 대한민국입니다. 이순신 장군은 한산도에서 싸웠다."
output = ner(text)
print(output)
```

> ℹ️ **transformers 5.x 대응**: 이 task의 `pipeline`은 transformers 5.0에서 제거되었습니다. 아래 코드는 `AutoModelFor...`를 직접 사용하도록 재작성했습니다. (transformers 5.3에서 동작 확인)

## question-answering

```python
# transformers 5.x에서 question-answering pipeline 제거 → AutoModelForQuestionAnswering 직접 사용
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
import torch

qa_name = "distilbert/distilbert-base-cased-distilled-squad"
qa_tokenizer = AutoTokenizer.from_pretrained(qa_name)
qa_model = AutoModelForQuestionAnswering.from_pretrained(qa_name)

question = "Where do I work?"
context = "My name is Sylvain and I work at todaycode in seoul"

inputs = qa_tokenizer(question, context, return_tensors="pt")
with torch.no_grad():
    outputs = qa_model(**inputs)

# 정답 구간: start/end 로짓의 argmax
start = outputs.start_logits.argmax()
end = outputs.end_logits.argmax() + 1
answer = qa_tokenizer.decode(inputs["input_ids"][0][start:end])
print("answer:", answer)
```

> ℹ️ **transformers 5.x 대응**: 이 task의 `pipeline`은 transformers 5.0에서 제거되었습니다. 아래 코드는 `AutoModelFor...`를 직접 사용하도록 재작성했습니다. (transformers 5.3에서 동작 확인)

## summarization
* [sshleifer/distilbart-cnn-12-6 · Hugging Face](https://huggingface.co/sshleifer/distilbart-cnn-12-6)

```python
# transformers 5.x에서 summarization pipeline 제거 → AutoModelForSeq2SeqLM 직접 사용
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

sum_name = "sshleifer/distilbart-cnn-12-6"
sum_tokenizer = AutoTokenizer.from_pretrained(sum_name)
sum_model = AutoModelForSeq2SeqLM.from_pretrained(sum_name)

text = """
    America has changed dramatically during recent years. Not only has the number of
    graduates in traditional engineering disciplines such as mechanical, civil,
    electrical, chemical, and aeronautical engineering declined, but in most of
    the premier American universities engineering curricula now concentrate on
    and encourage largely the study of engineering science. As a result, there
    are declining offerings in engineering subjects dealing with infrastructure,
    the environment, and related issues, and greater concentration on high
    technology subjects, largely supporting increasingly complex scientific
    developments. While the latter is important, it should not be at the expense
    of more traditional engineering.

    Rapidly developing economies such as China and India, as well as other
    industrial countries in Europe and Asia, continue to encourage and advance
    the teaching of engineering. Both China and India, respectively, graduate
    six and eight times as many traditional engineers as does the United States.
    Other industrial countries at minimum maintain their output, while America
    suffers an increasingly serious decline in the number of engineering graduates
    and a lack of well-educated engineers.
"""

inputs = sum_tokenizer(text, return_tensors="pt", truncation=True)
summary_ids = sum_model.generate(**inputs, max_length=130, min_length=30, num_beams=4)
print(sum_tokenizer.decode(summary_ids[0], skip_special_tokens=True))
```

* [gogamza/kobart-summarization · Hugging Face](https://huggingface.co/gogamza/kobart-summarization)

```python
# kobart 요약: 전용 토크나이저(PreTrainedTokenizerFast) 사용
# 주의: AutoTokenizer는 이 모델에서 RobertaTokenizer를 잘못 로드 → 출력이 깨짐. PreTrainedTokenizerFast 명시.
from transformers import PreTrainedTokenizerFast, AutoModelForSeq2SeqLM

ko_sum_name = "gogamza/kobart-summarization"
ko_sum_tokenizer = PreTrainedTokenizerFast.from_pretrained(ko_sum_name)
ko_sum_model = AutoModelForSeq2SeqLM.from_pretrained(ko_sum_name)

ko_text = "상위 몇 개의 높은 확률을 띠는 토큰을 출력할지 top_k 인자를 통해 조절합니다. 여기서 모델이 특이한 <mask> 단어를 채우는 것을 주목하세요. 이를 마스크 토큰(mask token)이라고 부릅니다. 다른 마스크 채우기 모델들은 다른 형태의 마스크 토큰을 사용할 수 있기 때문에 다른 모델을 탐색할 때 항상 해당 모델의 마스크 단어가 무엇인지 확인해야 합니다. 위젯에서 사용되는 마스크 단어를 보고 이를 확인할 수 있습니다."

input_ids = ko_sum_tokenizer.encode(ko_text, return_tensors="pt")
summary_ids = ko_sum_model.generate(
    input_ids, max_length=64, min_length=15,
    num_beams=4, no_repeat_ngram_size=2, eos_token_id=ko_sum_tokenizer.eos_token_id
)
print(ko_sum_tokenizer.decode(summary_ids[0], skip_special_tokens=True))
```

> ℹ️ **transformers 5.x 대응**: 이 task의 `pipeline`은 transformers 5.0에서 제거되었습니다. 아래 코드는 `AutoModelFor...`를 직접 사용하도록 재작성했습니다. (transformers 5.3에서 동작 확인)

## translation

```python
# transformers 5.x에서 translation pipeline 제거 → AutoModelForSeq2SeqLM 직접 사용
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

fr_name = "Helsinki-NLP/opus-mt-fr-en"
fr_tokenizer = AutoTokenizer.from_pretrained(fr_name)
fr_model = AutoModelForSeq2SeqLM.from_pretrained(fr_name)

inputs = fr_tokenizer("Ce cours est produit par Hugging Face.", return_tensors="pt")
translated = fr_model.generate(**inputs)
print(fr_tokenizer.decode(translated[0], skip_special_tokens=True))
```
