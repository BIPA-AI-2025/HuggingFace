## 3. Named Entity Recognition

- 문장 속에서 특정한 의미를 가진 단어(=개체명)를 찾아내고, 그 종류(=엔티티의 유형)**를 분류하는 자연어처리(NLP) 기술

- NER 시스템은 보통 아래와 같은 유형으로 분류합니다:

    PERSON : 사람 이름

    ORGANIZATION : 조직, 회사, 기관

    LOCATION : 지명, 장소

    DATE : 날짜, 시간

    MONEY :	금액

    EVENT :	사건, 행사

 - The full list of models for text classification 

   https://huggingface.co/models?pipeline_tag=token-classification&sort=downloads

```python
ner_classifier = pipeline(
    model="dslim/bert-base-NER-uncased", aggregation_strategy="simple"
)
```

```python
sentence = "I like to travel in Montreal."
entity = ner_classifier(sentence)
print(entity)
```

'entity_group': 'LOC'	추출된 단어는 위치(Location), 즉 지명임을 의미합니다. LOC = location.

'score': 0.9976744	모델이 이 단어가 지명일 확률을 99.77%로 예측했다는 의미입니다. (자신감 점수)

'word': 'montreal'	추출된 개체명은 montreal입니다. 모델이 소문자로 예측했지만 실제 문장에서는 대문자 Montreal로 쓰였습니다.

'start': 20	문장에서 시작 인덱스입니다. 20번째 문자부터 시작하며, Python 인덱스 기준입니다.

'end': 28	문장에서 끝 인덱스입니다. 28번째 문자 전까지 (end는 exclusive) 포함합니다. 즉 "Montreal"은 sentence[20:28]로 추출됩니다.

## Text Generation

- The full list of models for text classification 

  https://huggingface.co/models?pipeline_tag=text-generation

```python
text_generator = pipeline(model="gpt2")
```

```python
prompt = "If it is sunny today then "
```

```python
text_generator(prompt, do_sample=False)
```

```python
# 텍스트 생성
output = text_generator(prompt, do_sample=False, max_new_tokens=50)
```

```python
generated_text = output[0]['generated_text']
print("Prompt:")
print(prompt)
print("\n Generated Continuation:")
print(generated_text[len(prompt):])  # 프롬프트 제외하고 생성된 부분만 출력

print("\n Full Generated Text:")
print(generated_text)  # 전체 출력 (프롬프트 + 생성된 문장)
```

### 다양한 설정 적용

- max_new_tokens	새로 생성할 최대 토큰 수 :50 (기본), 100, 200 등
- temperature	창의성 조절 (낮을수록 보수적, 높을수록 다양함) : 1.0 (기본), 0.7 (더 보수적), 1.5 (더 창의적)
- top_k	상위 k개 후보만 고려 (값이 작을수록 안정적)	: 50, 100, 0 (제한 없음)
- top_p	누적 확률이 p 이상인 후보만 고려 (nucleus sampling)	:  0.9, 0.8, 1.0
- do_sample	True면 샘플링 (랜덤), False면 greedy (가장 확률 높은 것 선택) :	True 또는 False

```python
# 텍스트 생성
output = text_generator(
    prompt,
    max_new_tokens=50,
    temperature=0.9,   # 다양성 조절
    top_k=50,          # 상위 50개 후보 중에서 샘플링
    top_p=0.95,        # 누적 확률 95% 이상에서만 샘플링
    do_sample=True     # 랜덤 샘플링 사용
)
```

```python
# 출력 정리
generated_text = output[0]['generated_text']
print("Prompt:")
print(prompt)
print("\n Generated Continuation:")
print(generated_text[len(prompt):])  # 생성된 부분만 출력

print("\n Full Generated Text:")
print(generated_text)
```

## Text Translation

- Other advanced models  

  https://huggingface.co/models?pipeline_tag=translation

```python
# 번역 파이프라인 생성
en_fr_translator = pipeline("translation_en_to_fr", model="t5-small")
```

```python
# 입력 문장
text = "Hi, How are you?"
```

```python
# 번역 수행
output = en_fr_translator(text, max_length=40, clean_up_tokenization_spaces=True)
```

```python
# 결과 출력
print(" Original (EN):")
print(text)

print("\n Translated (FR):")
print(output[0]['translation_text'])

print("\n Full Output Dictionary:")
print(output[0])
```

### 한국어 → 영어 번역 (KO → EN)

```python
from transformers import pipeline

# 한국어 → 영어 번역 파이프라인 생성
ko_en_translator = pipeline("translation", model="Helsinki-NLP/opus-mt-ko-en")

# 번역할 문장
text_ko = "저는 주말에 한국 영화를 보는 것을 좋아해요."

# 번역 수행
output = ko_en_translator(text_ko, max_length=100)

# 결과 출력
print("원문 (KO):", text_ko)
print("번역 (EN):", output[0]['translation_text'])
```
